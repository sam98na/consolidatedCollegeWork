```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
###   RANDOM NUM GENERATOR   ###
def cons_region_generator(length = 10000):
    return_list = []
    for i in range(length):
        low = np.random.uniform(-10, 0)
        high = np.random.uniform(0, 11)
        return_list.append((low, high))
    return np.array(return_list)
```

# Question 1

```python
sample_list = cons_region_generator()
summed_ones = sum([1/(i[1]-i[0]) for i in sample_list if i[0] <= 1 <= i[1]])
summed_heights = sum([1/(i[1]-i[0]) for i in sample_list])
print("The probability of x = 1 being in the concept given x = 0 is is "\
      + str(summed_ones/summed_heights))
```

# Question 2

```python
def probability_finder(x, iterations = 5, length = 10000):
    probs = []
    for i in range(iterations):
        cons_region = cons_region_generator(length)
        summed_ones = sum([1/(i[1]-i[0]) for i in cons_region if i[0] <= x <= i[1]])
        summed_heights = sum([1/(i[1]-i[0]) for i in cons_region])
        probs.append(summed_ones/summed_heights)
    return np.average(probs)

xvals = np.linspace(-10, 10, 200)
yvals = []
for i in xvals:
    yvals.append(probability_finder(i))
plt.plot(xvals, yvals)
plt.xlabel("X Values")
plt.ylabel("Probabilities")
plt.title("Probabilities of Different X Values Being In Consequential Regions")
```

<b>This function when plotted looks like an inverted triangle, with the highest probability at 0 and dropping off at either end to tend towards 0. This makes sense because the further away from the stimulus (aka x=0) we are, the probability of it being in the region also drops. </b>

# Problem 4

```python
xvals = np.linspace(-5, 5, 100)
yvals = []
for i in xvals:
    probability = probability_finder(i)
    if probability == 0:
        yvals.append(0)
    else:
        yvals.append(np.log(probability_finder(i)))
plt.plot(xvals, yvals)
plt.xlabel("X Values")
plt.ylabel("Logged Probabilities")
plt.title("Logged Probabilities of Different X Values Being In Consequential Regions (X Range = -5, 5)")
```

```python
xvals = np.linspace(-10, 10, 200)
yvals = []
for i in xvals:
    probability = probability_finder(i)
    if probability == 0:
        yvals.append(0)
    else:
        yvals.append(np.log(probability_finder(i)))
plt.plot(xvals, yvals)
plt.xlabel("X Values")
plt.ylabel("Logged Probabilities")
plt.title("Logged Probabilities of Different X Values Being In Consequential Regions (X Range = -10, 10)")
```

<b>Using the reasoning from question 3, we can interpret these graphs to be showing that from the x range -5 to 5 has the probabilities decrease exponentially (as when an exponentiated curve is logged, it becomes a straight line). From the graph that shows the range (-10, -5][5, 10), we see that the graph curves, meaning whilst the probabilities are still decreasing, it is doing so in a non exponential way. We can interpret this as meaning probabilities drop the most when they start to move away from the stimulus, and mellows out in it's descent the further away it gets from the stimulus. </b>
